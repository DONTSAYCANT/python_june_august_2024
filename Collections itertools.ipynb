```python
import collections

#Advanced DS operations
```

```python
#Counter

Li1 = [5,6,7,4,3,4,5,5,5,4,4,3,2,1,1,2,3,6,4]

print(collections.Counter(Li1))
```

```python
a = "letter"

print(collections.Counter(a))
```

```python
#ChainMap

a = {1:"python",2:"cpp",3:"java"}
b = {6:"js", 0:"c"}
d = {1:"python",2:"cpp",3:"java"}

c = collections.ChainMap(a,b,d,h)
print(c)
```

```python
#Defaultdict

a = {}
a[3] = 30
a[2] = 10
a[5] = 50

print(a)
print(a[4])
```

```python
a = collections.defaultdict(int)
a[3] = 30
a[2] = 10
a[5] = 50

print(a)
print(a[4])
```

```python
#Deque

b = collections.deque() #Advanced List
print(b)
```

```python
b.appendleft(30)
print(b)

b.appendleft(60)
print(b)

b.append(60)
print(b)
#appendleft
#extendleft
#popleft
```

```python
#NamedTuple

Student = collections.namedtuple("Student", "Name Rollno Marks")

stud1 = Student("Sandy", 121, 88)

print(stud1)

print(stud1.Name)

print(stud1[0])
```

```python
val = ("Sandy", 121, 88)

print(val[1])
```

```python
#Ordereddict

a = {}

a[1] = 10
a[2] = 20
a[3] = 30
print(a)

a.pop(2)

print(a)

a[2] = 20

print(a)

a[1] = 100

print(a)
```

```python
a = collections.OrderedDict()

a[1] = 10
a[2] = 20
a[3] = 30
print(a)

a.pop(2)

print(a)

a[2] = 20

print(a)

a[1] = 100

print(a)
```

```python
## Itertools


#permutation and combination

import itertools
```

```python
Li1 = [2,3,4,1,5,6,7,8,9,10,11]
```

```python
val = list(itertools.permutations(Li1))
```

```python
print(val)
```

```python
val = list(itertools.combinations(Li1,4))
```

```python
print(val)
```

```python
val = list(itertools.combinations(Li1,3))
```

```python
print(val)
```
